```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt
```

```python
X_clustering = mall_customers[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(X_clustering)
```

```python
kmedoids = KMedoids(n_clusters=5, random_state=42)
kmedoids_labels = kmedoids.fit_predict(X_clustering)
```

```python
plt.figure(figsize=(14, 6))
```

```python
plt.subplot(1, 2, 1)
plt.scatter(X_clustering['Annual Income (k$)'], X_clustering['Spending Score (1-100)'], c=kmeans_labels, cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.colorbar()
```

Visualisasi hasil clustering menggunakan K-Means dengan 5 kluster pada data pelanggan mall. Masing-masing kluster diwakili oleh warna berbeda dalam plot, menggambarkan grup pelanggan berdasarkan pendapatan tahunan mereka dan skor pengeluaran.

```python
plt.subplot(1, 2, 2)
plt.scatter(X_clustering['Annual Income (k$)'], X_clustering['Spending Score (1-100)'], c=kmedoids_labels, cmap='viridis')
plt.title('K-Medoids Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.colorbar()
```

Visualisasi hasil clustering menggunakan K-Medoids dengan 5 kluster pada data pelanggan mall. Masing-masing kluster diwakili oleh warna berbeda dalam plot, menggambarkan grup pelanggan berdasarkan pendapatan tahunan mereka dan skor pengeluaran.
